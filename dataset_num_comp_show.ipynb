```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def autolabel(rects, ax):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.01*height,
                '%d' % int(height),
                ha='center', va='bottom')
```

```python
plt.rcParams.update({'font.size': 13})

fig,ax = plt.subplots(2, 1, figsize=(16,10))

dataset = ['DAQUAR', 'VQA Abs v2.0', 'TextVQA', 'Visual7W', 'VQA Abs', 'COCO-QA', 'CLEVR', 'VisualGenome', 'GQA', 'VQA', 'VQA v2.0']
y_pos = np.arange(len(dataset))
imcnt = [1449, 15623, 28408, 47300, 50000, 69172, 100000, 108077, 113018, 204721, 204721]

rects1 = ax[0].bar(y_pos, imcnt, align='center', alpha=0.5, color='r')
ax[0].set_xticks(y_pos)
ax[0].set_xticklabels(dataset)
ax[0].set_ylabel('#Images')
ax[0].set_ylim([0,imcnt[-1]*1.1])
# ax[0].set_title('视觉问答数据集图像数量比较')

dataset = ['DAQUAR', 'VQA Abs v2.0', 'TextVQA', 'COCO-QA', 'VQA Abs', 'Visual7W', 'VQA', 'CLEVR', 'VQA v2.0', 'VisualGenome', 'GQA']
y_pos = np.arange(len(dataset))
qacnt = [12468, 33383, 45336, 117684, 150000, 327939, 614163, 864968, 1105904, 1773258, 22669678]

rects2 = ax[1].bar(y_pos, qacnt, align='center', alpha=0.5, color='g')
ax[1].set_xticks(y_pos)
ax[1].set_xticklabels(dataset)
ax[1].set_ylabel('#QA pairs')
ax[1].set_ylim([0,qacnt[-1]*1.1])
# ax[1].set_title('视觉问答数据集问答对数量比较')

autolabel(rects1, ax[0])
autolabel(rects2, ax[1])

plt.show()
```
