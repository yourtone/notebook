```python
import json
jsondata = json.load(open('/home/lyt/.visdom/lyt_201907_28194431.json'))
data = jsondata['jsons']['window_37935b3dbccb7c']['content']['data']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.subplots(figsize=(10, 7))
for i in range(2):
    x = data[i]['x']
    y = data[i]['y']
    name = data[i]['name']
    plt.plot(x, y, label=name, linewidth=2)
    plt.title("Training loss and Learning Rate", fontsize=20)
    plt.xlabel("Epochs", fontsize=16)
#     plt.ylabel("Accuracy", fontsize=10)
    plt.tick_params(axis='both', labelsize=12)
    plt.grid(True, linestyle = "dashed", color = "gray", linewidth = "1")
plt.axis([0, int(max(data[0]['x'])+1), 0, max(data[0]['y'])*1.05])
plt.legend(loc='best', fontsize=14)
# plt.show()
plt.savefig('train_loss_and_lr.eps', format='eps', dpi=1000)
```

```python
plt.subplots(figsize=(10, 7))
for i in range(2,4):
    x = data[i]['x']
    y = data[i]['y']
    name = data[i]['name']
    plt.plot(x, y, label=name, linewidth=2)
    plt.title("Training and Validation Accuracies", fontsize=20)
    plt.xlabel("Epochs", fontsize=16)
    plt.ylabel("Accuracy", fontsize=16)
    plt.tick_params(axis='both', labelsize=12)
    plt.grid(True, linestyle = "dashed", color = "gray", linewidth = "1")
plt.axis([0, int(max(data[2]['x'])+1), 0, max(data[2]['y'])+1])
plt.legend(loc='best', fontsize=14)
# plt.show()
plt.savefig('train_and_val_acc.eps', format='eps', dpi=1000)
```
