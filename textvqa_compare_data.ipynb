```python
import numpy as np
```

```python
ori_ansemb = np.load('/home/data3/lyt/ocr_4/ocr_bert.new.768/trval/itoa_emb.npy')
# new_ansemb = np.load('/home/data3/lyt/concept_vqa/dataTVQA/image-feature/ocr/bert.768.0init.ori/TVA/ans_emb.npy')
new_ansemb = np.load('/home/data3/lyt/concept_vqa/dataTVQA/image-feature/ocr/bert.768.0init/TVA/ans_emb.npy')
```

```python
cnt=0
ind=0
oa_stack, na_stack = [], []
oaE_stack, naE_stack = [], []
for oa,na,oaE,naE in zip(ori_itoa, new_itoa, ori_ansemb, new_ansemb):
    if oa!=na:
        print('{}:{}|{}'.format(ind,oa,na))
        cnt+=1
        if len(oa_stack)==0 or len(na_stack)==0:
            oa_stack.append(oa)
            na_stack.append(na)
            oaE_stack.append(oaE)
            naE_stack.append(naE)
        else:
            if oa_stack[-1]==na:
                _ = oa_stack.pop()
                oaEs = oaE_stack.pop()
                assert((oaEs==naE).all())
            else:
                na_stack.append(na)
                naE_stack.append(naE)
            if na_stack[-1]==oa:
                _ = na_stack.pop()
                naEs = naE_stack.pop()
                assert((naEs==oaE).all())
            else:
                oa_stack.append(oa)
                oaE_stack.append(oaE)
    else:#oa==na
        assert (oaE==naE).all() or oa=='', '({})'.format(ind)
    ind+=1
print(cnt)
print(len(oa_stack))
print(len(na_stack))
```
