```python
import json
import numpy as np
```

```python
root_folder = '/home/lyt/code/vqa-concept'
data_folder = '{}/dataTVQA'.format(root_folder)
```

```python
#=======================================
```

```python
split_name = 'train'
```

```python
im_ocr = json.load(open('{}/ocr_{}.json'.format(data_folder, split_name),'r'))
ocrchars = {}
for item in im_ocr.values():
    ocrs = item['ocr_tokens']
    for ocr in ocrs:
        for c in ocr:
            if c in ocrchars:
                ocrchars[c] += 1
            else:
                ocrchars[c] = 1
```

```python
len(ocrchars)
```

```python
sorted_ocrchars = sorted(ocrchars.items(), key=lambda x: x[1], reverse=True)
print([item[0] for item in sorted_ocrchars])
```

```python
# assert all common chars in ocrchars
for i in range(ord('0'), ord('9')+1):
    if chr(i) not in ocrchars:
        print(chr(i))
for i in range(ord('A'), ord('Z')+1):
    if chr(i) not in ocrchars:
        print(chr(i))
for i in range(ord('a'), ord('z')+1):
    if chr(i) not in ocrchars:
        print(chr(i))
```
