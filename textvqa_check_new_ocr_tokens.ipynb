```python
split_name = 'train'
if split_name == 'test':
    img_folder = img_folder_test
jsdata = json.load(open('{}/TextVQA_0.5_{}.assemble.json'.format(data_folder, split_name),'r'))
print(jsdata.keys())
```

```python
data = jsdata['data']
print(len(data),type(data))
```

```python
pair = np.random.choice(data)
if len(pair['ocr_info']) != len(pair['ocr_tokens']):
    print(pair['ocr_tokens'])
```

```python
ocr_info = pair['ocr_info']
j = 0
word = ocr_info[-1]['word']
bbox = ocr_info[-1]['bounding_box']
img = imread('{}/{}.jpg'.format(img_folder, pair['image_id']))
print(img.shape)
print(word)
fig,ax = plt.subplots(figsize=(12, 12))
ax.imshow(img)
ax.add_patch(Rectangle((round(bbox['top_left_x']*pair['image_width']),round(bbox['top_left_y']*pair['image_height'])),
                       round(bbox['width']*pair['image_width']),round(bbox['height']*pair['image_height']),
                       linewidth=1,edgecolor='r',facecolor='none'))
plt.show()
```
